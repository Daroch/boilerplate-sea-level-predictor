```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
df = pd.read_csv('epa-sea-level.csv')
df
```

```python
#df = df[df['Year'] >= 2000]
#df = df.dropna()
x = df['Year']
y = df['CSIRO Adjusted Sea Level']
fig, ax = plt.subplots(1, 1, figsize=(9,3))
ax.set_xlabel('Year')
ax.set_ylabel('Sea Level (inches)')
ax.set_title('Rise in Sea Level')
```

```python
res = linregress(x, y)
plt.plot(x, y,label='original data')
plt.plot(x, res.intercept + res.slope*x, 'r', label='fitted line')
plt.xlim(1850, 2075)
#ax.set_xticks(range(1850, 2075, 25))
plt.legend()
```

```python
final_point = 2050
x2 = range(1880, final_point)
y2 = res.intercept + res.slope*x2
res2 = linregress(x2, y2)
plt.plot(x2, y2,label='original data')
plt.plot(x2, res2.intercept + res2.slope*x2, 'r', label='fitted line')
plt.xlim(1850, final_point)
plt.legend()

```
